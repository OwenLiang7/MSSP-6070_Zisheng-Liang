```python
import pandas as pd

file_path = '/content/drive/MyDrive/MSSP 6070_Zisheng Liang/c. Modules/Week4/parsed.csv'
df = pd.read_csv(file_path)

# check first 5 rows
df.head()
```

a. Find the 95th percentile of earthquake magnitude in Japan using magType of 'mb'

```python
jp_mb = df[(df['place'].str.contains('Japan', na=False)) & (df['magType'] == 'mb')]
jp_95th = jp_mb['mag'].quantile(0.95)
print("95th percentile of 'mb' earthquakes in Japan:", jp_95th)
```

b. Percentage of earthquakes in Indonesia that were coupled with tsunamis

```python
indo = df[df['place'].str.contains('Indonesia', na=False)]
tsunami_pct = (indo['tsunami'] == 1).sum() / len(indo) * 100
print(" Percentage of Indonesia earthquakes with tsunamis:", round(tsunami_pct, 2), "%")
```

c. Summary statistics for earthquakes in Nevada

```python
nevada = df[df['place'].str.contains('Nevada', na=False)]
print(" Descriptive statistics for earthquakes in Nevada:")
print(nevada.describe())
```

d. Add column indicating Ring of Fire membership

```python
# List of Ring of Fire locations
ring_of_fire_places = [
    'Bolivia', 'Chile', 'Ecuador', 'Peru', 'Costa Rica', 'Guatemala', 'Mexico',
    'Japan', 'Philippines', 'Indonesia', 'New Zealand', 'Antarctica',
    'Canada', 'Fiji', 'Alaska', 'Washington', 'California', 'Russia',
    'Taiwan', 'Tonga', 'Kermadec'
]

# Function to check if place belongs to Ring of Fire
def is_ring_of_fire(place):
    if pd.isnull(place):
        return False
    for loc in ring_of_fire_places:
        if loc in place:
            return True
    return False

# Add the column
df['ring_of_fire'] = df['place'].apply(is_ring_of_fire)

# Check preview
df[['place', 'ring_of_fire']].head()
```

e. Count earthquakes in and outside Ring of Fire

```python
ring_count = df['ring_of_fire'].sum()
non_ring_count = len(df) - ring_count

print(" Number of earthquakes in Ring of Fire:", ring_count)
print(" Number of earthquakes outside Ring of Fire:", non_ring_count)
```

f. Count of tsunami events in Ring of Fire

```python
ring_tsunamis = df[(df['ring_of_fire']) & (df['tsunami'] == 1)]
print(" Number of tsunami events in Ring of Fire:", len(ring_tsunamis))
```
